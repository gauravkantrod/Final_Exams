<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#db0b49;">3. Read the dataset(GlobalTemperatures.csv) and Answer below questions?
    
    
   1. Set the date column as a DateTimeIndex and sort it?
   2. Resample annually and rename index & columns?
   3. Calculate the global baseline temperature?
   4. Create the new "temperature anomaly" column (['AnnualAverageTemp'] - global_ref_temp)?
   5. Drop the "AnnualAverageTemp" because you created new column name "Temperature Anomaly"?

```python
temp = pd.read_csv('GlobalTemperatures.csv')
```

```python
temp.head()
```

```python
temp.tail()
```

```python
temp.dtypes
```

```python
100*temp.isnull().sum()/temp.shape[0]
```

#### 1. Set the date column as a DateTimeIndex and sort it?

```python
temp['DateTimeIndex'] = pd.DatetimeIndex(temp['dt'])
```

```python
temp['DateTimeIndex'].head()
```

#### 2. Write a code to convert from yyyy-mm-dd to yyyy?

```python
temp['year'] = temp['dt'].apply(lambda x : x.split('-')[0])
```

```python
temp['year'].head()
```

#### 3.Calculate the global baseline temperature on (1951 to 1980)

```python
temp['year'] = temp['year'].astype(int)
```

```python
temp_1951_1980 = temp[(temp['year'] >= 1952) & (temp['year'] <= 1980)]
```

```python
global_ref_temp = temp_1951_1980['LandAndOceanAverageTemperature'].sum ()/temp_1951_1980.shape[0]
```

```python
global_ref_temp = round(global_ref_temp,2)
global_ref_temp
```

#### 4. Create the new "temperature anomaly" column (['AnnualAverageTemp'] - global_ref_temp)?

```python
temp['temperature_anomaly'] = temp['LandAndOceanAverageTemperature'] - global_ref_temp
```

#### 5. Drop the "AnnualAverageTemp" becuse you created new column name "Temperature Anomaly"?

```python
temp.drop('LandAndOceanAverageTemperature', axis = 1, inplace = True)
```
